Popularity based recommendation system...(on the basis of ratings)

Merging two datasets

```python
ratings_merge_books=ratings.merge(books,on="ISBN")
ratings_merge_books
```

```python
ratings_merge_books.shape
```

```python
book_rating_df=ratings_merge_books.groupby("Book-Title").count()["Book-Rating"].reset_index()
book_rating_df
```

```python
avg_rating_df=ratings_merge_books.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_rating_df.rename(columns={"Book-Rating":"Average-Rating"},inplace=True)
avg_rating_df
```

```python
popular_df=book_rating_df.merge(avg_rating_df,on="Book-Title")
popular_df
```

```python
popular_df=popular_df[popular_df["Book-Rating"]>=250].sort_values("Average-Rating",ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-M","Book-Rating","Average-Rating"]]
```

Collaborative filtering based recommendation system...

Filtering on users and books:Users who have rated 200 books and books with ratings > 50

```python
import pandas as pd
books=pd.read_csv("Books(271361 records).csv")
users=pd.read_csv("Users.csv")
ratings=pd.read_csv("Ratings.csv")
```

```python
books
```

```python
users
```

```python
ratings
```

```python
ratings_with_books=ratings.merge(books,on="ISBN")
```

```python
ratings_with_books
```

```python
x=ratings_with_books.groupby("User-ID").count()["Book-Rating"]>200
x[x]
```

```python
x[x].index
```

```python
sincere_users=x[x].index
```

```python
filtered_rating=ratings_with_books[ratings_with_books["User-ID"].isin(sincere_users)]
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
y[y].index
```

```python
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
pt=final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
import numpy as np
```

Function to recommend

```python
def recommend(book_name):
  index=np.where(pt.index==book_name)[0][0]
  similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
  for i in similar_items:
    print(pt.index[i[0]])
print("Similar books recommendation:")
recommend("Year of Wonders")
```
